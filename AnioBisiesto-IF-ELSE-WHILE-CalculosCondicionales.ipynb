```python
n_1 = float(input("Introduce un número: ") )
n_2 = float(input("Introduce otro número: ") )
opcion = 0
           
print("¿Qué operación desea hacer? \n1) Sumar los dos números\n2) Restar los dos números\n3) Multiplicar los dos números")
opcion = int(input("Introduce un número para la operación: ") )     

if opcion == 1:
    print("La suma de",n_1,"+",n_2,"es",n_1+n_2)
elif opcion == 2:
    print("La resta de",n_1,"-",n_2,"es",n_1-n_2)
elif opcion == 3:
    print("El producto de",n_1,"*",n_2,"es",n_1*n_2)
else:
    print("Opción incorrecta")
```

# Año Bisiesto + Godoy

```python

while True:
  try:
    fecha = int(input("Introduzca un año para verificar si es biciesto, "))
  except ValueError: #Es para asegurarnos que sólo escriba un número y no un str
        print("Debes escribir un número.")
        continue
  if fecha !=0 and fecha % 4 != 0:
    print("No es bisiesto, ingresar 0 para salir")
  elif fecha !=0 and fecha % 4 == 0 and fecha % 100 != 0: 
	  print("Es bisiesto, ingresar 0 para salir")
  elif fecha % 4 == 0 and fecha % 100 == 0 and fecha % 400 != 0: 
	  print("No es bisiesto, ingresar 0 para salir")
  elif fecha !=0 and fecha % 4 == 0 and fecha % 100 == 0 and fecha % 400 == 0:  
	  print("Es bisiesto, ingresar 0 para salir")
  if fecha == 0: #Salir del bucle
    print("Programa terminado")
    break
```
