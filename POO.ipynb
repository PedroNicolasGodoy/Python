```python
class Animal:
  def __init__(self, especie, edad) -> None:
      self.especie=especie
      self.edad=edad

  def caminar(self):
    pass
  def comer(self):
    pass
  def hablar(self):
    pass
  def describeme(self):
    return f" Soy {self.especie} y tengo {self.edad} años de edad. {type(self).__name__}"

class Domestico(Animal):
  pass

class Perro(Domestico):
  def __init__(self, especie, edad, nombre) -> None:
      super().__init__(especie, edad)
      self.nombre = nombre
  def caminar(self):
    print("camino con mis cuatro patas")
  def comer(self):
    print("que rico!!")
  def hablar(self):
    print("Guau, Guau!!")
  def describeme(self):
    cadena_padre=super().describeme()
    return cadena_padre + f" Mi nombre es {self.nombre}"


perro=Perro("Perro", 8 , "Lasi")
print(perro.describeme())
```
