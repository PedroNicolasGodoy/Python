```python
def area_rectangulo(base, altura):
  return base * altura

area_rectangulo(15, 10)
```

```python
import math

def area_circulo (radio):
  return math.pi * radio**2

area_circulo(5)
```

```python
def relacion(num1, num2):
  if num1 > num2:
    return 1
  elif num1 < num2:
    return -1
  elif num1 == num2:
    return 0

print(relacion(5, 10), relacion(10, 5), relacion(5, 5))
```

```python
def intermedio (num1, num2):
  return (num1 + num2)/2

print(intermedio (-12, 24))
```

FunciÃ³n recortar con n+umero inf y numero sup

```python
def recortar(num, lim_inf, lim_sup):
  return max(min(num, lim_sup), lim_inf)

recortar(15, 0, 10)
```

FunciÃ³n para separa pares e impares

```python
numeros=[6,5,2,1,7]

def separar(*args):
  lista = sorted(args)
  pares = []
  impares = []
  for arg in lista:
    if arg % 2 == 0:
      pares.append(arg)
    else:
      impares.append(arg)      
  return pares, impares

pares, impares = separar(*numeros)
print(pares)
print(impares)
```
