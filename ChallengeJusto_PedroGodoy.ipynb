```python
# Obtener las columnas relevantes para la comparación
df1 = clap.MONTO
df2 = bansur.MONTO

# Comparar el valor pagado entre los dos DataFrames
df_comparacion = pd.merge(clap, bansur, on='TARJETA', how='inner')
df_comparacion['Diferencia'] = df1 - df2
df_comparacion = df_comparacion[(df_comparacion['Diferencia'].abs() <= 0.99) | (df1.equals(df2))]

print(df_comparacion)
```
